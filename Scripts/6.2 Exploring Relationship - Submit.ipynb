```python
# create a subset excluding unneeded columns

sub = df[['sales', 'quantity', 'discount', 'profit', 'shipping_cost']]
```

### Exploring relationships

#### Correlations:

```python
# Create a correlation matrix using pandas
sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

There is a strong relationship (0.77) between sales and shipping cost.
There is a weak relationship (0.27) between quantity and shipping cost.
There is a moderate relationship (0.48) between sales and profit.
There is a moderate relationship (0.31) between sales and quantity.
There is a weak relationship (-0.087) between sales and discount.
